```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for consistency
np.random.seed(10)

# Simulate stock prices for 100 days
days = np.arange(1, 101)
stock_A = np.cumsum(np.random.randn(100) * 2 + 100)  # Stock A
stock_B = np.cumsum(np.random.randn(100) * 1.5 + 120)  # Stock B
stock_C = np.cumsum(np.random.randn(100) * 1.8 + 90)  # Stock C

# Combine stocks into a NumPy array
stocks = np.vstack([stock_A, stock_B, stock_C])

# Labels for stocks
stock_labels = ['Stock A', 'Stock B', 'Stock C']

# Display first few data points
print("Stock Data Shape:", stocks.shape)  # (3, 100)
print("Sample Data:\n", stocks[:, :5])  # First 5 days
```

```python
# Mean price per stock
mean_prices = np.mean(stocks, axis=1)

# Standard deviation (volatility)
volatility = np.std(stocks, axis=1)

# Daily returns (percentage change)
daily_returns = np.diff(stocks) / stocks[:, :-1] * 100  # (3, 99)



# Overall stock growth (percentage)
growth = ((stocks[:, -1] - stocks[:, 0]) / stocks[:, 0]) * 100

# Print results
for i, stock in enumerate(stock_labels):
    print(f"{stock}: Mean Price = {mean_prices[i]:.2f}, Volatility = {volatility[i]:.2f}, Growth = {growth[i]:.2f}%")
    # print(i,stock)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(days, stock_A, label="Stock A", linestyle="--", marker="o")
plt.plot(days, stock_B, label="Stock B", linestyle=":", marker="s")
plt.plot(days, stock_C, label="Stock C", linestyle="-.", marker="d")

plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.title("Stock Market Simulation")
plt.legend()
plt.grid()
plt.show()
```
